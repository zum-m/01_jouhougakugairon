```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'col1':[1,2], "col2": [3,4]})



height_list = [185,162,171,155,191,166]
weight_arr = np.array([72,51,69,55,87,78])

height_series = pd.Series(height_list)
weight_series = pd.Series(weight_arr)


print(height_series)
print(weight_series)
```

indexラベルをつける


```python
labels = ["A","B","C","D","E","F"]
height_series = pd.Series(height_list, index= labels)
weight_series = pd.Series(weight_arr, index=labels)



print(height_series)
print(weight_series)
```

辞書がたからインデックス付きのシリーズ作成

```python
height_dic = {"a":185,"b":162,"c":171,"d":155,"e":191,"f":166}
height_series = pd.Series(height_dic)

print(height_series)
```

```python
age_series = pd.Series([19,19,20,18,22,21],index = labels)
height_series = pd.Series(height_list, index= labels)
weight_series = pd.Series(weight_arr, index=labels)


table_df = pd.DataFrame({"height":height_series,"weight":weight_series,"age":age_series,})

print(table_df)
```

dfの値を参照する




```python
# カラム指定してインデックス参
table_df[["height","weight"]]
# table_df[[],["A","C"]]
```

```python
# 比較演算子で欲しい行を取る

table_df[table_df['age']>=20]
```

```python
# ロケーションのメソッド（ラベルを使う」）で特定の場所をとる
table_df.loc["C",:]
```

```python
# インデックスでロケーションを指定して参照
table_df.iloc[2,2]
table_df.iloc[2:,1:]
```

```python
# データフレームの値の変更
print(table_df)
table_df.iloc[2,2] = 21

print(table_df)
```

```python
# データフレームの列のついか
table_df['sex'] = ["m","f","m","f","m","f"]
table_df
```

```python
# 行の追加
val  = [160,52,20,"m"]

val = pd.Series(val,index=["height","weight","age","sex"],name='G')
table_df.append(val)
```
