```python
import numpy as np

x = np.arange(1, 11)
y1 = np.repeat(3, 10)
y2 = np.repeat(2, 10)
y3 = np.repeat(1, 10)
markers1 = [".", ",", "o", "v", "^", "<", ">", "1", "2", "3"]
markers2 = ["4", "8", "s", "p", "*", "h", "H", "+", "x", "D"]
markers3 = ["d", "|", "_", "None", None, "", "$x$", "$\\alpha$", "$\\beta$", "$\\gamma$"]
for i in x-1:
    plt.scatter(x[i], y1[i], s=100, marker=markers1[i])
    plt.scatter(x[i], y2[i], s=100, marker=markers2[i])
    plt.scatter(x[i], y3[i], s=100, marker=markers3[i])
```
