````python
from sklearn.datasets import load_iris

# あやめデータセットを読み込む
iris = load_iris()

# データセットの特徴量（説明変数）を取得
# X = iris.data
X = iris.data[:,2:]
# print(X)

# データセットのターゲット（目的変数）を取得
y = iris.target

# データセットの特徴量名を取得
feature_names = iris.feature_names

# データセットのターゲット名を取得
target_names = iris.target_names
````

````python
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans

wcss = []  # これ何？Within-Cluster Sum of Squares　クラスタ内のデータ点とそのクラスタの重心との距離の二乗和を表します。このリストは、後でWCSSの値を格納するために使用されます。

for i in range(1,11): # for文の中身でやってることがわからないので具体的に理解したい
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('The wlbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

````

WCSSにはどのような値が入っていき、プロットされるのですか？

ChatGPT
WCSSは、クラスタ内のデータ点とそのクラスタの重心との距離の二乗和を表す値です。WCSSの値は、クラスタ数を変化させたときのクラスタリングモデルの性能を示す指標となります。

このコードでは、1から10までの範囲のクラスタ数に対してK-meansクラスタリングを実行し、各クラスタ数におけるWCSSの値を計算しています。ループの中で、kmeans.inertia_がWCSSの値を返し、それがwcssリストに追加されます。

最終的に、plt.plot()関数を使用して、クラスタ数とWCSSの値を折れ線グラフとしてプロットします。横軸にクラスタ数、縦軸にWCSSをとり、クラスタ数が増えるにつれてWCSSがどのように変化するかを可視化します。

グラフのプロット結果を見ることで、WCSSの値がどのように変化するかを把握することができます。エルボー法として知られる手法では、WCSSのグラフが肘（エルボー）のような形状をしており、WCSSの減少が緩やかになる部分が最適なクラスタ数の目安とされます。



```
for i in range(1, 11):  # for文の中身でやっていることは何？
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```
次の行からのforループでは、クラスタ数の範囲を指定しています（1から10）。ループの中では、KMeansクラスのインスタンスを作成し、n_clustersパラメータにクラスタ数を指定します。kmeans.fit(X)は、Xを特徴量としてK-meansクラスタリングを実行し、クラスタリングモデルを学習します。その後、kmeans.inertia_はWCSSの値を返し、wcssリストに追加されます。

````python
kmeans = KMeans(n_clusters=3)
y_pred = kmeans.fit_predict(X)

plt.scatter(X[y_pred == 0,0],X[y_pred ==0,1],s = 100,c='red')
plt.scatter(X[y_pred == 1,0],X[y_pred ==1,1],s = 100,c='blue')
plt.scatter(X[y_pred == 2,0],X[y_pred ==2,1],s = 100,c='green')


plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s = 100,c= 'yellow',label = 'Centroids')
plt.legend()
````
