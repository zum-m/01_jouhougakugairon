```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data
Y = iris.target

# # 標準化
scaler = StandardScaler()
Xstd = scaler.fit_transform(X)


```

```python
# 主成分分析（PCA
from sklearn.decomposition import PCA
# 二次元二次元削減
pca = PCA(n_components=2)
X_pca = pca.fit_transform(Xstd)
```

```python
# 次元削減後の可視化
import matplotlib.pyplot as plt
import numpy as np

colors = ['r','b','g']
markers = ['s','x','o']

for i,c,m in zip(np.unique(Y),colors,markers):
    plt.scatter(X_pca[Y==i,0],X_pca[Y==i,1],c=c,label = i ,marker = m)
    
    
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.legend(loc= 'lower right')
plt.show()
```

IOSMAP

```python
from sklearn.manifold import Isomap

iso = Isomap(n_components=2)
X_iso = iso.fit_transform(Xstd)


import matplotlib.pyplot as plt
import numpy as np


colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']

for i, c, m in zip(np.unique(Y), colors, markers):
    plt.scatter(X_iso[Y == i, 0], X_iso[Y == i, 1], c=c, label=i, marker=m)


plt.xlabel('pc1')
plt.ylabel('pc2')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(Xstd)


colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']

for i, c, m in zip(np.unique(Y), colors, markers):
    plt.scatter(X_tsne[Y == i, 0], X_tsne[Y == i, 1], c=c, label=i, marker=m)


plt.xlabel('pc1')
plt.ylabel('pc2')
plt.legend(loc='lower right')
plt.show()
```
