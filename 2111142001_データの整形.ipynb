### 複数のデータフレームwオブジェクトを連結(concat)

```python
# 辞書を使って、カラムラベルを指定したDataFrameオブジェクトを生成
df1 = pd.DataFrame({
    "a":["a0","a1","a2"],
    "b":["b1","b2","b3"],
    "c":["c1","c2","c3"]
})

df2 = pd.DataFrame({
    "a":["a3","a4","a5"],
    "b":["b3","b4","b5"],
    "c":["c3","c4","c5"]
})

print(df1)
print(df2)
```

```python
# 列方向にデータフレームを連結
out_0 = pd.concat([df1,df2],axis=0)
print(out_0)


# インデックスラベルを付け直す
out_0 = pd.concat([df1,df2],axis=0,ignore_index= True)
print(out_0)
```

```python
# 行方向に連結
out_1 = pd.concat([df1,df2],axis= 1)

out_1
```

```python
# カラムラベルと連結の関係
# カラムラベルが一致しないと？
df3 = pd.DataFrame({
    "a":["a3","a4","a5"],
    "b":["b3","b4","b5"],
    "d":["c3","c4","c5"]
})

print(df3)
out_tmp = pd.concat([df1,df3],axis=0)
print(out_tmp)
```

```python
# 内部結合と外部結合
inner = pd.concat([df1,df1],axis=0,join='inner')
print(inner)

outer = pd.concat([df1,df3],axis=0,join='outer')
print(outer)
```

```python
# まーじ：複数sのデータに含まれている特定のカラムをもとに一つにまとめる操作（）merge
leftdf = pd.DataFrame({
    "a":["a0","a1","a2"],
    "b":["b1","b2","b3"],
    "key":["k0","k1","k2"]
})

rightdf= pd.DataFrame({
    "c":["c0","c1","c2"],
    "d":["d0","d1","d2"],
    "key":["k0","k1","k2"]

})
testdf= pd.DataFrame({
    "c":["c0","c1","c2"],
    "d":["d0","d1","d2"],
    "key":["k1","k2","k0"]

})

df_merge = pd.merge(leftdf,rightdf, on='key')
print(df_merge)
df_merge2 = pd.merge(leftdf,testdf, on='key')
print(df_merge2)
```

```python
height_list = [185, 162, 171, 155, 191, 166]
label_1 = ['A','B','C','D','E','F']
weight_list = [72, 51, 69, 55, 87, 78]
label_2 = ['A', 'B', 'C','D','G','H']

height_df = pd.DataFrame(pd.Series(height_list, index=label_1))
weight_df = pd.DataFrame(pd.Series(weight_list, index= label_2))
df1 = pd.merge (height_df, weight_df, left_index=True, right_index=True, how='outer')
print (df1)
df2 = pd.merge (height_df, weight_df, left_index=True, right_index=True, how='inner')
print (df2)
```
